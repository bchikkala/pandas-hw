```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
load_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(load_file)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
purchase_data_dup = purchase_data_df["SN"].drop_duplicates()

total_players_count = pd.DataFrame({
    "The Total Number of Players": [purchase_data_dup.count()]
})

count = purchase_data_dup.count()
total_players_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_dup = purchase_data_df["Item ID"].drop_duplicates()
unique = item_dup.count()
avg_price = purchase_data_df["Price"].mean()
total_purchases = purchase_data_df["Price"].count()
total_revenue = purchase_data_df["Price"].sum()

summary_df = pd.DataFrame({
    "Total Number of Unique Items": [unique],
    "Average Price": [purchase_data_df["Price"].mean()],
    "Total Number of Purchases": [purchase_data_df["Price"].count()],
    "Total Revenue": [purchase_data_df["Price"].sum()]
})
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_dup_df = purchase_data_df[["Gender", "SN", "Age"]]
player_dup_df = player_dup_df.drop_duplicates()
player_gender = player_dup_df["Gender"].value_counts()

percent_players = (player_gender / count) * 100

gender_dem_df = pd.DataFrame({
    "Total Count": player_gender,
    "Percent of Players": percent_players
})
gender_dem_df
```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
age_range = [ "<10", "10-14", "15-19","20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Total Count"] = pd.cut(purchase_data_df["Age"],
                                        bins, labels=age_range)
age_groups = purchase_data_df.groupby("Total Count")
age_count = age_groups.count()

organized_age_summary = age_count[["Age"]]
organized_age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_SN = purchase_data_df.groupby("SN")

spender_count = spender_SN["Purchase ID"].count()

avg_spender_price = spender_SN["Price"].mean()

spender_total = spender_SN["Price"].sum()

spender_peak = pd.DataFrame({
    "Purchase Count": spender_count,
    "Average Purchase Price": avg_spender_price,
    "Total Purchase Price": spender_total})

format_peak = spender_peak.sort_values([
    "Total Purchase Price"], ascending=False).head()
format_peak["Average Purchase Price"] = format_peak["Average Purchase Price"].map("${:,.2f}".format)
format_peak["Total Purchase Price"] = format_peak["Total Purchase Price"].map("${:,.2f}".format)
format_peak.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item = purchase_data_df[["Item ID", "Item Name", "Price"]]

item_info = item.groupby(["Item ID", "Item Name"])

item_count = item_info["Price"].count()

item_value = (item_info["Price"].sum())

item_price = item_value / item_count

item_pop = pd.DataFrame({
    "Purchase Count": item_count,
    "Item Price": item_price,
    "Total Purchase Value": item_value
})
item_pop_format = item_pop.sort_values([
    "Purchase Count"], ascending=False)
item_pop_format["Item Price"] = item_pop_format["Item Price"].map("${:,.2f}".format)
item_pop_format["Total Purchase Value"] = item_pop_format["Total Purchase Value"].map("${:,.2f}".format)
item_pop_format.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_pop_format = item_pop.sort_values([
    "Total Purchase Value"],
    ascending=False)
item_pop_format["Item Price"] = item_pop_format["Item Price"].map("${:,.2f}".format)
item_pop_format["Total Purchase Value"] = item_pop_format["Total Purchase Value"].map("${:,.2f}".format)
item_pop_format.head()
```
